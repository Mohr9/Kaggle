```python
import numpy as np
import pandas as pd

# Création de variables aléatoires linéaires
np.random.seed(0)
X_linear = np.linspace(0, 10, 100)
y_linear = 2 * X_linear + np.random.randn(100) * 2  # Relation linéaire y = 2*X + bruit

# Création du DataFrame
df_linear = pd.DataFrame({'X': X_linear, 'y': y_linear})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Supposons que 'X' est votre variable continue et 'y' est votre variable binaire
sns.scatterplot(x='X', y='y', data=df_linear)
plt.xlabel('Variable Continue')
plt.ylabel('Variable Binaire')
plt.title('Relation entre la variable continue et la variable binaire')
plt.show()
```

```python
# Création de variables aléatoires non linéaires
np.random.seed(0)
X_nonlinear = np.linspace(-5, 5, 100)
y_nonlinear = X_nonlinear**2 + np.random.randn(100)  # Relation non linéaire y = X^2 + bruit

# Création du DataFrame
df_nonlinear = pd.DataFrame({'X': X_nonlinear, 'y': y_nonlinear})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Supposons que 'X' est votre variable continue et 'y' est votre variable binaire
sns.scatterplot(x='X', y='y', data=df_nonlinear)
plt.xlabel('Variable Continue')
plt.ylabel('Variable Binaire')
plt.title('Relation entre la variable continue et la variable binaire')
plt.show()
```
