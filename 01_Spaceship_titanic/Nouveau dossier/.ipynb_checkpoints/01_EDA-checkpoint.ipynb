# Dataset Description

* In this competition your task is to predict whether a passenger was transported to an alternate dimension during the Spaceship Titanic's collision with the spacetime anomaly. To help you make these predictions, you're given a set of personal records recovered from the ship's damaged computer system.

- File and Data Field Descriptions
* train.csv - Personal records for about two-thirds (~8700) of the passengers, to be used as training data.

* test.csv - Personal records for the remaining one-third (~4300) of the passengers, to be used as test data. Your task is to predict the value of Transported for the passengers in this set.
* sample_submission.csv - A submission file in the correct format.
    - PassengerId - Id for each passenger in the test set.
    - Transported - The target. For each passenger, predict either True or False.

#### Observation :

- Taille des boxplot similaire en fonction de la catégories pour chaque variable : **égalité des variances**
- ici on voit mal les médianes pour chaque variables, s'il y a peu de variation de la médiane alors il y a de faible chance qu'il y ait une dépendance, cependant nous allons faire un test statistique adapté pour en savoir plus
##### toutes les conditions ici sont réunis pour faire un test student (t-test) afin de vérifier s'il existe une dépendance entre notre target et les variables continues ou non

```python
#### Création d'une fonction renvoyant le résultat d'un test student entre deux variables : 

def ttest(data,var_qualitative, var_continue):
    # Test de Student pour comparer les moyennes des autres variables
    statistic, p_value = ttest_ind(data[var_qualitative], data[var_continue])
    alpha = 0.05
    
    
    print(f"Test de Student :  \n{var_continue} \nStatistique = {statistic}, p-value = {p_value}")

    if p_value < alpha:
        print("p-value ≤ α , il y a une différence/corrélation entre les populations.")
    else:
        print("p-value ≥ α , il n’y a pas de différence/relation entre les populations de l’étude.")
        #Ce sont les variables qui ne seront pas conservées
        return var_a_suppr
```

- **Valeur de p ≤ α :** les variables présentent une association statistiquement significative (rejeter H0)
    * Si la valeur de p est ≤ au seuil de signification, vous pouvez rejeter l'hypothèse nulle et conclure qu'il existe une association statistiquement significative entre les variables.
- **Valeur de p > α :** impossible de conclure que les variables sont associées (ne pas rejeter H0)
    * Si la valeur de p est > au seuil de signification, vous ne pouvez pas rejeter l'hypothèse nulle, car vous ne possédez pas suffisamment de preuves pour conclure que les variables sont associées.


```python
var_a_suppr = [(ttest(df, target.name, i)) for i in var_num]
```

```python
var_a_suppr = [i for i in var_a_suppr if i!=None]
print("Liste des variables a supprimer :", var_a_suppr)
```

On voit qu'aucune variable a des valeurs > à alpha, ainsi, la liste des variables a supprimer est vide, donc nous allons **conserver toutes les variables numériques**

#### Test statistiques Variable catégoriques Vs catégoriques

- Maintenant, nous allons effectuer un test khi2 et vérifier s'il existe ou non une dépendance entre ces variables

```python
 a = pd.crosstab(df.Transported, df.RoomService)
```

```python
def khi2(data, target, var_qualitative):
    # Test de Student pour comparer les moyennes des autres variables
    df_temp = pd.crosstab(data[target], data[var_qualitative])
    p_val = chi2_contingency(df_temp)[1]
    alpha = 0.05
    
    print(f"Test khi2 :  \n{var_qualitative} \np-value = {p_val.round(3)}")

    if p_val < alpha:
        print("p-value ≤ α , il y a une différence/corrélation entre les populations.")
    else:
        print("p-value ≥ α , il n’y a pas de différence/relation entre les populations de l’étude.")
        #Si c'est cette condition qui ait rempli, alors la variable ne sera pas conservée et sera supprimée
        return  var_qualitative
```

```python
var_a_suppr = [khi2(df, target.name, i) for i in var_cat]
var_a_suppr = [var for var in var_a_suppr if var is not None]  # Filtrer les valeurs None
var_a_suppr
```

##  Cercle corrélation : une autre manière de  visualiser la relation variables initiales - variables synthétiques

```python
def correlation_graph(pca, 
                      x_y, 
                      features) : 
    """Affiche le graphe des correlations
    Positional arguments : 
    -----------------------------------
    pca : sklearn.decomposition.PCA : notre objet PCA qui a été fit
    x_y : list ou tuple : le couple x,y des plans à afficher, exemple [0,1] pour F1, F2
    features : list ou tuple : la liste des features (ie des dimensions) à représenter
    """
    # Extrait x et y 
    x,y=x_y
    # Taille de l'image (en inches)
    fig, ax = plt.subplots(figsize=(8,7))
    # Pour chaque composante : 
    for i in range(0, pca.components_.shape[1]):
        # Les flèches
        ax.arrow(0,0, 
                pca.components_[x, i],  
                pca.components_[y, i],  
                head_width=0.07,
                head_length=0.07, 
                width=0.02, )
        # Les labels
        plt.text(pca.components_[x, i] + 0.05,
                pca.components_[y, i] + 0.05,
                features[i])      
    # Affichage des lignes horizontales et verticales
    plt.plot([-1, 1], [0, 0], color='grey', ls='--')
    plt.plot([0, 0], [-1, 1], color='grey', ls='--')

    # Nom des axes, avec le pourcentage d'inertie expliqué
    plt.xlabel('F{} ({}%)'.format(x+1, round(100*pca.explained_variance_ratio_[x],1)))
    plt.ylabel('F{} ({}%)'.format(y+1, round(100*pca.explained_variance_ratio_[y],1)))
    plt.title("Cercle des corrélations (F{} et F{})".format(x+1, y+1))
   
    # Le cercle 
    an = np.linspace(0, 2 * np.pi, 100)
    plt.plot(np.cos(an), np.sin(an))  # Add a unit circle for scale
    # Axes et display
    plt.axis('equal')
    plt.show(block=False)
```

```python
correlation_graph(pca1, (0,1), X_std.columns)
correlation_graph(pca1, (2,3), X_std.columns)
```

### Visualisation PCA n dimensions avec clusterisation KMeans :

```python
def kmeans_elbow(data, nclust):
    visualizer = KElbowVisualizer(estimator=KMeans(),k=(1, nclust))
    visualizer.fit(data)
    visualizer.show()
```

```python
kmeans_elbow(X_pca, nclust=10)
```

5 Cluster semble être le meilleur compromis

```python
def graph_kmean(data, n_clust, x,y):
    """data : dataframe a analyser
    x,y : numéro des composantes à analyser """
    #Entrainement :
    kmeans = KMeans(n_clusters=n_clust)
    kmeans.fit(data)
    labels = kmeans.labels_
    new_lab = [f"Cluster {i}" for i in labels]
    #Visualisation :
    sns.reset_defaults()
    sns.scatterplot(data = data, x=data[:,x], y = data[:,y], hue = new_lab, hue_order= np.unique(new_lab), 
                    palette="bright")
    plt.xlabel(f"F{x+1}") #+1 car on commence a compter à partir de F1 et non F0
    plt.ylabel(f"F{y+1}")#meme raisonnement
    plt.show()
```

#### Réduisons le nombre de cardinalité de la variable Age pour la transformer en variable ordinale :

```python
def map_age(x):
    if x<1:
        return "baby"
    elif (x>=1)&(x<=10):
        return "kids"
    elif (x>10)&(x<=18):
        return "teenagers"
    elif (x>18&x<75):
        return "adult"
    else:
        return "Aged"
```

```python
df['Age'] = df["Age"].map(map_age)
```
